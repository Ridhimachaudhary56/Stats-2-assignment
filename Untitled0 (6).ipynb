```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-score: {z_score}")
    print(f"P-value: {p_value}")
    if p_value < alpha:
        print("Result: Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
    else:
        print("Result: Fail to reject the null hypothesis (H0). No significant difference between the sample and population mean.")
sample_data = [51, 52, 50, 49, 53, 54, 52, 51, 50, 53]
population_mean = 50
population_std = 2

z_test(sample_data, population_mean, population_std)
```

```python
import numpy as np
import scipy.stats as stats
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    margin_of_error = stats.t.ppf((1 + confidence) / 2, df=len(sample) - 1) * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
ci_lower, ci_upper = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
def z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    n = len(sample)
    standard_error = population_std / np.sqrt(n)
    z_score = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_value
def t_test(sample, population_mean):
    t_stat, p_value = stats.ttest_1samp(sample, population_mean)
    return t_stat, p_value
np.random.seed(42)
population_mean = 50
population_std = 10
large_sample = np.random.normal(loc=population_mean, scale=population_std, size=100)
small_sample = np.random.normal(loc=population_mean, scale=population_std, size=15)
z_score, z_p_value = z_test(large_sample, population_mean, population_std)
t_stat, t_p_value = t_test(small_sample, population_mean)
print("Z-Test Results:")
print(f"  Z-score: {z_score:.4f}, P-value: {z_p_value:.4f}")
print("\nT-Test Results:")
print(f"  T-statistic: {t_stat:.4f}, P-value: {t_p_value:.4f}")
plt.figure(figsize=(8, 5))
plt.bar(['Z-Test', 'T-Test'], [z_score, t_stat], color=['blue', 'red'], alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.title("Comparison of Z-Test and T-Test Results")
plt.ylabel("Test Statistic")
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
def margin_of_error(sample, confidence_level=0.95):
    n = len(sample)
    if n < 2:
        return "Sample size too small to calculate MOE."

    mean = np.mean(sample)
    std_dev = np.std(sample, ddof=1)
    alpha = 1 - confidence_level
    z_score = stats.norm.ppf(1 - alpha/2)

    moe = z_score * (std_dev / np.sqrt(n))

    return moe
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
moe_95 = margin_of_error(sample_data, confidence_level=0.95)
print(f"Margin of Error (95% Confidence Level): {moe_95:.4f}")
moe_99 = margin_of_error(sample_data, confidence_level=0.99)
print(f"Margin of Error (99% Confidence Level): {moe_99:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    P_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * (1 - prior_A))
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / P_B

    return posterior_A_given_B
prior_disease = 0.01
likelihood_positive_given_disease = 0.95
likelihood_positive_given_no_disease = 0.05
posterior = bayes_theorem(prior_disease, likelihood_positive_given_disease, likelihood_positive_given_no_disease)
print(f"Probability of having the disease given a positive test result: {posterior:.4f}")
priors = np.linspace(0.001, 0.1, 100)
posteriors = [bayes_theorem(p, 0.95, 0.05) for p in priors]
plt.figure(figsize=(8, 5))
plt.plot(priors, posteriors, label="Posterior Probability", color="blue")
plt.axhline(y=0.5, linestyle="--", color="red", label="50% Threshold")
plt.xlabel("Prior Probability of Disease")
plt.ylabel("Posterior Probability Given Positive Test")
plt.title("Effect of Prior Probability on Posterior")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Function to calculate expected frequencies for a Chi-square test.

    Parameters:
    observed (2D list or NumPy array): Observed frequency table

    Returns:
    expected (NumPy array): Expected frequencies table
    """
    # Convert observed data to a NumPy array
    observed = np.array(observed)

    # Compute row and column sums
    row_totals = observed.sum(axis=1, keepdims=True)  # Sum along rows
    col_totals = observed.sum(axis=0, keepdims=True)
    grand_total = observed.sum()
    expected = (row_totals @ col_totals) / grand_total

    return expected
observed_data = [
    [30, 10],
    [15, 25]
]
df_observed = pd.DataFrame(observed_data, index=['Men', 'Women'], columns=['Coffee', 'Tea'])
expected_frequencies = calculate_expected_frequencies(observed_data)
df_expected = pd.DataFrame(expected_frequencies, index=['Men', 'Women'], columns=['Coffee', 'Tea'])
print("Observed Frequencies:")
print(df_observed)
print("\nExpected Frequencies:")
print(df_expected)
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

def check_anova_assumptions(groups):
    print("Checking ANOVA Assumptions...\n")
    for i, group in enumerate(groups, start=1):
        stat, p_value = stats.shapiro(group)
        print(f"Shapiro-Wilk Test (Normality) for Group {i}: W = {stat:.4f}, p = {p_value:.4f}")
        if p_value < 0.05:
            print("Reject normality assumption (Data is NOT normally distributed)\n")
        else:
            print("Normality assumption holds (Data is normally distributed)\n")
    stat, p_value = stats.levene(*groups)
    print(f"Levene's Test (Equal Variance): W = {stat:.4f}, p = {p_value:.4f}")
    if p_value < 0.05:
        print("Reject equal variance assumption (Variances are NOT equal)\n")
    else:
        print("Equal variance assumption holds (Variances are equal)\n")
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)
group4 = np.random.normal(65, 10, 30)

check_anova_assumptions([group1, group2, group3, group4])
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_f_distribution(df1, df2, x_max=5, num_points=1000):
    """Plot the F-distribution with given degrees of freedom."""
    x = np.linspace(0, x_max, num_points)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, alpha=0.2, color='blue')
    plt.axvline(stats.f.ppf(0.95, df1, df2), color='red', linestyle='--', label='Critical Value (95%)')

    plt.title("F-Distribution")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

def perform_anova():
    """Perform a one-way ANOVA test and visualize results with boxplots."""
    np.random.seed(42)
    group1 = np.random.normal(loc=20, scale=5, size=30)
    group2 = np.random.normal(loc=22, scale=5, size=30)
    group3 = np.random.normal(loc=24, scale=5, size=30)

    f_stat, p_value = stats.f_oneway(group1, group2, group3)
    print(f"F-statistic: {f_stat:.2f}, p-value: {p_value:.4f}")

    data = pd.DataFrame({
        "Group": ["Group 1"] * 30 + ["Group 2"] * 30 + ["Group 3"] * 30,
        "Values": np.concatenate([group1, group2, group3])
    })

    plt.figure(figsize=(8, 5))
    sns.boxplot(x="Group", y="Values", data=data)
    plt.title("One-Way ANOVA Test Results")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

def main():
    df1 = 5
    df2 = 10
    plot_f_distribution(df1, df2)
    perform_anova()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def plot_f_distribution(df1, df2, x_max=5, num_points=1000):
    """Plot the F-distribution with given degrees of freedom."""
    x = np.linspace(0, x_max, num_points)
    y = stats.f.pdf(x, df1, df2)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, alpha=0.2, color='blue')
    plt.axvline(stats.f.ppf(0.95, df1, df2), color='red', linestyle='--', label='Critical Value (95%)')

    plt.title("F-Distribution")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()

def perform_anova():
    """Perform a one-way ANOVA test and visualize results with boxplots."""
    np.random.seed(42)
    group1 = np.random.normal(loc=20, scale=5, size=30)
    group2 = np.random.normal(loc=22, scale=5, size=30)
    group3 = np.random.normal(loc=24, scale=5, size=30)

    f_stat, p_value = stats.f_oneway(group1, group2, group3)
    print(f"F-statistic: {f_stat:.2f}, p-value: {p_value:.4f}")

    data = pd.DataFrame({
        "Group": ["Group 1"] * 30 + ["Group 2"] * 30 + ["Group 3"] * 30,
        "Values": np.concatenate([group1, group2, group3])
    })

    plt.figure(figsize=(8, 5))
    sns.boxplot(x="Group", y="Values", data=data)
    plt.title("One-Way ANOVA Test Results")
    plt.xlabel("Groups")
    plt.ylabel("Values")
    plt.grid()
    plt.show()

def perform_t_test():
    """Simulate random data from a normal distribution and perform a t-test."""
    np.random.seed(42)
    sample1 = np.random.normal(loc=20, scale=5, size=30)
    sample2 = np.random.normal(loc=22, scale=5, size=30)

    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    print(f"T-test statistic: {t_stat:.2f}, p-value: {p_value:.4f}")

def perform_variance_test():
    """Perform a hypothesis test for population variance using a Chi-square distribution."""
    np.random.seed(42)
    sample = np.random.normal(loc=20, scale=5, size=30)
    sample_var = np.var(sample, ddof=1)
    pop_var = 25

    chi_square_stat = (len(sample) - 1) * sample_var / pop_var
    p_value = stats.chi2.sf(chi_square_stat, df=len(sample)-1)

    print(f"Chi-square statistic: {chi_square_stat:.2f}, p-value: {p_value:.4f}")

def perform_z_test():
    """Perform a Z-test for comparing proportions between two datasets."""
    np.random.seed(42)
    count1, size1 = 50, 200
    count2, size2 = 60, 220

    prop1, prop2 = count1 / size1, count2 / size2
    pooled_prop = (count1 + count2) / (size1 + size2)
    se = np.sqrt(pooled_prop * (1 - pooled_prop) * (1/size1 + 1/size2))

    z_stat = (prop1 - prop2) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    print(f"Z-test statistic: {z_stat:.2f}, p-value: {p_value:.4f}")

def perform_f_test():
    """Perform an F-test for comparing variances of two datasets."""
    np.random.seed(42)
    sample1 = np.random.normal(loc=20, scale=5, size=30)
    sample2 = np.random.normal(loc=22, scale=7, size=30)

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_stat = var1 / var2 if var1 > var2 else var2 / var1
    df1, df2 = len(sample1) - 1, len(sample2) - 1
    p_value = 2 * (1 - stats.f.cdf(f_stat, df1, df2))

    print(f"F-test statistic: {f_stat:.2f}, p-value: {p_value:.4f}")

def perform_chi_square_gof():
    """Perform a Chi-square test for goodness of fit."""
    np.random.seed(42)
    observed = np.random.randint(10, 50, size=5)
    expected = np.full(5, np.mean(observed))

    chi2_stat, p_value = stats.chisquare(observed, expected)
    print(f"Chi-square statistic: {chi2_stat:.2f}, p-value: {p_value:.4f}")

    categories = [f"Category {i+1}" for i in range(5)]

    plt.figure(figsize=(8, 5))
    plt.bar(categories, observed, label='Observed', alpha=0.6, color='blue')
    plt.bar(categories, expected, label='Expected', alpha=0.6, color='red')
    plt.title("Chi-square Goodness of Fit Test")
    plt.xlabel("Categories")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid()
    plt.show()

def main():
    df1 = 5
    df2 = 10
    plot_f_distribution(df1, df2)
    perform_anova()
    perform_t_test()
    perform_variance_test()
    perform_z_test()
    perform_f_test()
    perform_chi_square_gof()

if __name__ == "__main__":
    main()
```
